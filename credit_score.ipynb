handeling the Payment_Behaviour

```python
#we generally need five main factors in detemining the credit score of an individuals
#credit mix:- it refers ato the diversity of the credit mix a individual have
# that is it should have credit card,loans.
# length of credit history it means how much time the individuals have the access to the 
# money that he/she have taken from the bank and use it
#  
credit_score_classification['Payment_Behaviour'].unique()
credit_score_classification.head()
```

```python
def text_preprocess(text):
    token = text.split("_")
    if(token[0] == 'high'):
        return 0
    else:
        return 1
```

```python
#we can assign value to 
#high:0, low:1
#small:0 medium:1 large:2
credit_score_classification['Payment_Behaviour'] = credit_score_classification['Payment_Behaviour'].apply(text_preprocess)
```

Annual income

```python
#handling the monthly inhand salary of the customers
def text_analysis(text):
    token = text.split("_")
    return token[0]
credit_score_classification['Annual_Income'] = credit_score_classification['Annual_Income'].apply(text_analysis)

#chaning the data type of annual income
credit_score_classification['Annual_Income'] = credit_score_classification['Annual_Income'].astype(float)
```

```python
credit_score_classification['Monthly_Inhand_Salary'] = credit_score_classification['Annual_Income'] / 12 +250
```

```python
#dropping the annual columns
credit_score_classification = credit_score_classification.drop(['Annual_Income'],axis = 1)
```

delay from the due date

```python
#handeling the delay_from_due_date 
#we can group them according to certain requirment
#we can bucket them as per our requirement

#we can assign the negative int with 'greater number' since these have paid there amount
  #before the due date:4
#then people who have taken more than 2 months can be assigned with 0
#then people with in between 1-2 monts 1
#then more than half of the month is 2
#then half is 3

#function to execute this task
def group_due_dates(delays):
  if(delays <= 0):
    return 4
  elif(delays >= 60):
    return 0
  elif(delays >= 30):
    return 1
  elif(delays >= 15):
    return 2
  else:
    return 3

credit_score_classification['Delay_from_due_date'] = credit_score_classification['Delay_from_due_date'].apply(group_due_dates)
```

```python
credit_score_classification = credit_score_classification.dropna(subset=['Num_of_Delayed_Payment'])
```

```python
#we have seperated the dataset into two parts x & y
x = credit_score_classification.iloc[:,1:4]
y = credit_score_classification['Credit_Mix']
x.head()
```
